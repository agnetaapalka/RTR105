Dotā funkcija: y=(1+x)*exp(x)

No numpy bibliotēkas importējam mums nepieciešamo _exp_  un _linspace_ funkcijas ;

no matplotlib importējam _pyplot_ zīmēšanai:

```python
import sys
sys.path.append("/usr/lib/python2.7/dist-packages/numpy/")
sys.path.append("/usr/lib/python2.7/dist-packages/matplotlib/")
from numpy import exp, linspace
from matplotlib import pyplot as plt
```

x asij ar linspace piešķiram masīvu ar 100 punktiem ar vērtībām -6;0:

```python
x = linspace(-6, 0, 100)
```

y argumentu definējam kā doto funkciju un attiecīgi tālāk y2, y3, y4 un y5 definējam kā funkcijas izvirzījumu:

```python
y = (1+x)*exp(x)
y2 = 1+x
y3 = (1+x)*(1+x)
y4 = y3 + ((x**2)/2) + ((x**3)/2)
y5 = y4 + ((x**3)/6) + ((x**4)/3)
```

Ar _grid_ piešķiram zīmējumam gradējumu; xlabel nosaucam par x un ylabel par f(x), kā arī dodam grafikam nosaukumu _title_ "Funkcija..." ;

ar _plot_ zīmējam grafikus un _show_ tos attēlo:

```python
plt.grid()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Funkcija $(1+x)*exp(x)$")
plt.plot(x, y)
plt.plot(x,y2)
plt.plot(x,y3)
plt.plot(x,y4)
plt.plot(x,y5)                  
plt.show()
```
