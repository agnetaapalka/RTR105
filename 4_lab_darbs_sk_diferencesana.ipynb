# 4. laboratorijas darbs - skaitliskā diferencēšana

Dotā funkcija: y=(1+x)*exp(x)

No bibliotēkām _matplotlib_ importējam _pyplot_ zīmēšanai; no numpy importējam _linspace_ un _exp_:

```python
from matplotlib import pyplot as plt
from numpy import exp, linspace
```

Definējam savu funkciju un x piešķiram masīvu no -4;1 ar 11 punktiem

```python
def f(x):
    return exp(x)*(1+x)
x = linspace(-4,1,11)
y = f(x)
```

Uzzīmējam pamatfunkciju un izveidojam leģendu (atšifrējumu):

```python
legend = []
from matplotlib import pyplot as plt
plt.axis([-4, 1, -1, 2])
plt.grid()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Funkcija $(1+x)*exp(x)$ un tās atvasinājumi")
plt.plot(x,y,"k")
legend.append("$(1+x)*exp(x)$ - pamatfunkcija - viss ir savienots ar taisnām līnijām")
plt.plot(x,y,"ro")
legend.append("$(1+x)*exp(x)$ - tikai daži punkti")
plt.legend(legend, loc =3)
```

Norādām, cik garš masīvs:

```python
N = len(x)

deltax = x[1] - x[0]
der = []
```

Rēķinām un zīmējam klāt pirmo atvasinājumu:

```python
N = len(x)

deltax = x[1] - x[0]
der = []

for i in range(N):
    temp = (f(x[i] + deltax) - f(x[i])) / deltax
    der.append(temp)


legend = []
from matplotlib import pyplot as plt
plt.axis([-4, 1, -1, 2])
plt.grid()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Funkcija $(1+x)*exp(x)$ un tās atvasinājumi")
plt.plot(x,y,"k")
legend.append("$(1+x)*exp(x)$ - pamatfunkcija - viss ir savienots ar taisnām līnijām")
plt.plot(x,y,"ro")
legend.append("$(1+x)*exp(x)$ - tikai daži punkti")
plt.plot(x, der, "y")
legend.append("$(1+x)*exp(x)$ atvasinājums - viss ir savienots ar taisnām līnijām")
plt.plot(x, der,"go")
legend.append("$(1+x)*exp(x)$ atvasinājums - daži punkti")
plt.legend(legend, loc =3, framealpha = 0.3)
```

Rēķinām un zīmējam klāt otro atvasinājumu:

```python
dersec = []

for i in range(N-1):
    temp = (der[i+1] - der[i]) / deltax
    dersec.append(temp)
```

```python
legend = []
from matplotlib import pyplot as plt
plt.axis([-4, 1, -1, 4])
plt.grid()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Funkcija $(1+x)*exp(x)$ un tās atvasinājumi")
plt.plot(x,y,"k")
legend.append("$(1+x)*exp(x)$ - pamatfunkcija - viss ir savienots ar taisnām līnijām")
plt.plot(x,y,"ro")
legend.append("$(1+x)*exp(x)$ - tikai daži punkti")
plt.plot(x, der, "y")
legend.append("$(1+x)*exp(x)$ atvasinājums - viss ir savienots ar taisnām līnijām")
plt.plot(x, der,"go")
legend.append("$(1+x)*exp(x)$ atvasinājums - daži punkti")
plt.plot(x[0:N-1], dersec, "m")
legend.append("$(1+x)*exp(x)$ otrās kārtas atvasinājums, viss ir savienots ar taisnām līnijam")
plt.plot(x[0:N-1], dersec, "bo")
legend.append("$(1+x)*exp(x)$ otrās kārtas atvasinājums, daži punkti")
plt.legend(legend, loc =3, framealpha = 0.3)
```
