# 5. laboratorijas darbs - skaitliskā integrēšana

Dotā funkcija: y=(1+x)*exp(x)

No bibliotēkām _math_ importējam funkciju _exp_; no _numpy_ importējam _random_ gadījuma skaitļu ģenerēšanai; no _matplotlib_ importējam _pyplot_ zīmēšanai:

```python
from numpy import random
from math import exp
from matplotlib import pyplot as plt
```

N argumentam piešķiram punktu daudzumu, kuru izmantosim, lai aizpildītu laukumu:

```python
N = 7000
```

Definējam funkcijas apgbalu (maksimālo un minimālo x), kura laukumu rēķināsim un šajā apgabalā ģenerējam dotos N punktus:

```python
xmin = -4
xmax = 0
x = random.uniform(xmin,xmax,N)
```

Sadalam mūsu apgabalu 8 daļās un pārbaudām vai tajās dotie punkti izlīdzinās vienmērīgi:

```python
k = [0, 0, 0, 0, 0, 0, 0, 0]
for i in range(N):
    if x[i] < -3.5:
        k[0] = k[0] + 1
    elif x[i] < -3:
        k[1] = k[1] + 1
    elif x[i] < -2.5:
        k[2] = k[2] + 1
    elif x[i] < -2:
        k[3] = k[3] + 1
    elif x[i] < -1.5:
        k[4] = k[4] + 1
    elif x[i] < -1:
        k[5] = k[5] + 1
    elif x[i] < -0.5:
        k[6] = k[6] + 1
    elif x[i] < 0:
        k[7] = k[7] + 1
    else:
        k[8] = k[8] + 1
print(k)
```

Norādam maksimālo un minimālo y, kā arī ģenerējam šajā apgabalā dotos N punktus:

```python
ymin = -0.5
ymax = 1.5
y = random.uniform(ymin, ymax, N)
```

1) Zīmējam grafika pamatu <br>
2) Ņemot vērā, ka izvēlētajā apgabalā grafikam ir gan pozitīvas, gan negatīvas vērtībās, izveidojam divus ciklus (y<0 un y>0) <br>
_Ja punkts pieder grafikam, to krāsojam zaļu un tos skaitam, ja nepieder - krāsojam sarkanu_

```python
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Funkcija un tās integrālis (laukums starp funkciju un x ass)')

N1 = 0
N2 = 0
for i in range(N):
    if y[i] > 0:
        if y[i] < (1+x[i])*exp(x[i]):
            plt.plot(x[i],y[i],'go')
            N1 = N1 + 1
        elif y[i] > (1+x[i])*exp(x[i]):
            plt.plot(x[i], y[i],'ro')
    if y[i] < 0:
        if y[i] > (1+x[i])*exp(x[i]):
            plt.plot(x[i], y[i],'go')
            N2 = N2 + 1
        elif y[i] < (1+x[i])*exp(x[i]):
            plt.plot (x[i], y[i],'ro')
```

Pēc dotajām formulām aprēķinām zināmo laukumu un attiecīgi arī nezināmo, kas tad arī ir mūsu funkcijas (noteiktajā apgabalā) integrālis, to arī izprintējam:

```python
S_zinaamais = (ymax-ymin) * (xmax-xmin)
S_nezinaamais = (S_zinaamais * (N1-N2) / N)
print(S_nezinaamais)
```
